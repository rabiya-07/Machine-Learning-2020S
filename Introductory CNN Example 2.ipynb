```python
# Import TensorFlow library
import tensorflow as tf

# Import NumPy library
import numpy as np

# Import the Sequential model and Dense layer from Keras
from tensorflow import keras

# Create a Sequential model (a linear stack of layers)
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

# Compile the model, specifying the optimizer and loss function
model.compile(optimizer='sgd', loss='mean_squared_error')

# Define the input (independent variable) data as a NumPy array
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)

# Define the target (dependent variable) data as a NumPy array
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

# Train the model on the input and target data for a specified number of epochs
model.fit(xs, ys, epochs=5)

# Print the model's prediction for a new input value (7)
print(model.predict([7]))
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

# Custom callback to display loss, weights, and learning rate
class CustomCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if epoch % 50 == 0:  # Print every 50 epochs
            print(f"Epoch {epoch + 1}/{self.params['epochs']}, Loss: {logs['loss']}")
            weights, biases = self.model.layers[0].get_weights()
            print(f"Weights: {weights}, Biases: {biases}")
            print(f"Learning Rate: {self.model.optimizer.lr.numpy()}")

# Create a Sequential model
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

# Compile the model with a custom optimizer and learning rate
custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=custom_optimizer, loss='mean_squared_error')

# Define the input (independent variable) data as a NumPy array
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)

# Define the target (dependent variable) data as a NumPy array
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

# Train the model on the input and target data for a specified number of epochs
history = model.fit(xs, ys, epochs=150, callbacks=[CustomCallback()])

# Print the model's prediction for a new input value (7)
print(model.predict([7]))
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

# Custom callback to display loss, weights, and learning rate at each epoch
class CustomCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        weights, biases = self.model.layers[0].get_weights()
        learning_rate = self.model.optimizer.lr.numpy()
        loss = logs['loss']
        print(f"Epoch {epoch + 1}/{self.params['epochs']}, Loss: {loss:.4f}, Weights: {weights}, Biases: {biases}, Learning Rate: {learning_rate:.4f}")

# Create a Sequential model
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

# Compile the model with a custom optimizer and learning rate
custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=custom_optimizer, loss='mean_squared_error')

# Define the input (independent variable) data as a NumPy array
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)

# Define the target (dependent variable) data as a NumPy array
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

# Train the model on the input and target data for a specified number of epochs
history = model.fit(xs, ys, epochs=5, callbacks=[CustomCallback()])

# Print the model's prediction for a new input value (7)
print(model.predict([7]))
```
